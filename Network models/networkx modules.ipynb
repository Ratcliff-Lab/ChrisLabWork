```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
def Assignment (G, proplist, percentagesList):
    '''
    Input: 
    A graph representing the physical interactions where bacteria will play games
    list of dictionaries representing different kinds of bacteria playing the games and their unique properties
    list of percentages representing the initial starting proportions of each kind of bacteria
        if it does not add up to 1, the remaining cells are randomly placed as empty cells
        
    Output: 
    A graph where the nodes are randomly given bacterial properties based on the percentages and properties given. 
    '''
    
    #randomly selects a subsection of the notes and progressively apples different bacterial properties to each node. 

    nodesLeft = list(G.nodes()) #generates a list from each of the nodes in a graph

    for i,v in enumerate (proplist): #for each set of bacterial properties
        randNodes = sample(nodesLeft,int(len(G.nodes())*percentagesList[i])) #generate a random subsection of the nodes
        nodesLeft = [x for x in nodesLeft if x not in randNodes] #list containing all the unselected nodes
        
        nodeDict = {} #initiate a blank dict to carry the node + properties 
        for j in randNodes: #for each node in the current random subsection of the node
            nodeDict[j] = v #fill the dict with the node/property pair
        nx.set_node_attributes(G,nodeDict) #actually apply the properties to the nodes on the graph 
        
    if len(nodesLeft) != 0: #if there are additional nodes left after assignment, randomly give them properties
        nodeDict = {} 
        for k in nodesLeft: 
            nodeDict[k] = random.choice(proplist)
        nx.set_node_attributes(G,nodeDict)
    return (G)
```

```python
def showGraph(G, graphFunction):
    c_map =  (list((nx.get_node_attributes(G,'color')).values()))
    graphFunction(G,with_labels=True, node_color = c_map)
    
    plt.show()
    plt.close()
```

```python
def createAntagonist(G):
    '''
    Input: 
    Takes in a physical connectivity graph that has labeled nodes
    
    Output: 
    A graph representing antagonistic relationships between the bacteria that are connected
    '''
    
    D = nx.create_empty_copy(G.to_directed()) #creating an empty digraph that has the same nodes as G

    Antagonism = nx.get_node_attributes(G,'antagonism') #get a dictionary of all the antagonism
    for i in G: #iterating over all of the nodes in G
        if Antagonism[i] == 0:
            continue
        neigh = nx.all_neighbors(G,i) #obtaining all the neighbors of node i
        lists = [(i,j,Antagonism[i]) for j in neigh] #creating a list of tuples containing a new edge and the edge weight based on the antagonism (insert formula for antagonism here if we decide to do a more subtle version)
        D.add_weighted_edges_from(lists)

    return (D)
```

```python
def createCooperation(G):
    '''
    Input: 
    Takes in a physical connectivity graph that has labeled nodes
    
    Output: 
    A graph representing beneficial relationships between connected nodes
    '''
    
    D = nx.create_empty_copy(G.to_directed()) #creating an empty digraph that has the same nodes as G

    Cooperation = nx.get_node_attributes(G,'cooperation') #get a dictionary of all the antagonism
    for i in G: #iterating over all of the nodes in G
        if Cooperation[i] == 0:
            continue
        neigh = nx.all_neighbors(G,i) #obtaining all the neighbors of node i
        lists = [(i,j,Cooperation[i]) for j in neigh] #creating a list of tuples containing a new edge and the edge weight based on the antagonism (insert formula for antagonism here if we decide to do a more subtle version)
        D.add_weighted_edges_from(lists)

    return (D)
```

```python
#Example input: 
G = nx.grid_graph ([1,5,4])
proplist = [{'color' : 'blue','antagonism': 0,'cooperation': 3, 'type': 'cooperator'}, 
            {'color' : 'red','antagonism': 5,'cooperation': 0, 'type': 'antagonist'},
            {'color' : 'green','antagonism': 0,'cooperation': 0, 'type': 'empty'}]
percentagesList = [0.5,0.5]


#testing the functions
G = Assignment(nx.grid_graph ([1,5,5]), proplist, [0.4,0.4,0.2])
D = createAntagonist(G)
H = createCooperation(G)

showGraph(G, nx.draw_kamada_kawai)
showGraph(D,nx.draw)
showGraph(H, nx.draw)
```
